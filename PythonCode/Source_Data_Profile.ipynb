```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
df_IOT_data = pd.read_csv('./ABT_DE_Exercise_V2-main/Sample_Data/IOT_data.csv', index_col=0)
df_Subject_Accelerometer_data = pd.read_csv('./ABT_DE_Exercise_V2-main/Sample_Data/Subject_Accelerometer.csv', index_col=0)
df_Subject_Demographic = pd.read_csv('./ABT_DE_Exercise_V2-main/Sample_Data/Subject_Demographic.csv', index_col=0)
df_Subject_Program = pd.read_csv('./ABT_DE_Exercise_V2-main/Sample_Data/Subject_Program.csv', index_col=0)
```

```python
profile = ProfileReport(df_IOT_data, title="df_IOT_data Profiling Report" ,minimal=True)
profile.to_file("df_IOT_data.html")

profile_df_Subject_Accelerometer = ProfileReport(df_Subject_Accelerometer_data, 
                                                 title="df_Subject_Accelerometer Profiling Report" ,minimal=True)
profile_df_Subject_Accelerometer.to_file("df_Subject_Accelerometer.html")


profile_df_Subject_Demographic = ProfileReport(df_Subject_Demographic, 
                                                 title="df_Subject_Demographic Profiling Report" ,minimal=True)
profile_df_Subject_Demographic.to_file("df_Subject_Demographic.html")


profile_df_Subject_Program = ProfileReport(df_Subject_Program, 
                                                 title="df_Subject_Program Profiling Report" ,minimal=True)
profile_df_Subject_Program.to_file("df_Subject_Program.html")
```
